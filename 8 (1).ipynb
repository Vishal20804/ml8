```python
Ans:1
import pandas as pd
df=pd.read_excel('flight_price.xlsx')
df.shape
df.describe
```

```python
Ans:2
import pandas as pd
import matplotlib.pyplot as plt

# Plotting histogram
plt.figure(figsize=(4, 4))
plt.hist(df['Price'], bins=30, edgecolor='black')
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.grid(True)
plt.show()
```

```python
Ans:3
import pandas as pd

# Assuming you have already loaded the dataset into a DataFrame named 'df'

# Find the minimum and maximum prices
min_price = df['Price'].min()
max_price = df['Price'].max()

# Calculate the range of prices
price_range = max_price - min_price

# Display the results
print(f"The minimum price is: {min_price}")
print(f"The maximum price is: {max_price}")
print(f"The range of prices is: {price_range}")
```

Ans:7
Date or Month: Analyzing flight prices over time can reveal seasonal patterns and trends. Grouping the data by month or date and examining the average prices for each period can help identify any recurring trends.

Departure and Arrival Cities: Prices may vary based on the departure and arrival cities. Analyzing price trends for specific routes can provide insights into popular and expensive destinations.

Airline: Different airlines may have different pricing strategies. Analyzing how prices vary across different airlines can help identify trends related to carrier choices.

Day of the Week: Prices can fluctuate depending on the day of the week. Analyzing price trends based on weekdays and weekends can highlight any patterns related to cheaper or more expensive travel days.

Advance Booking Time: Investigating how prices change based on the advance booking time can reveal trends related to early or last-minute bookings.

Holidays and Events: Prices may be impacted during holidays, festivals, or special events. Analyzing the prices around these periods can identify any trends related to peak travel times.
